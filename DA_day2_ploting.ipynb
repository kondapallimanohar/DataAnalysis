```python
#ploot the scores of two students in five difference subject
#subject as x axis
#marks as y axis
student1=np.array([70,45,20,30,60])
student2=np.array([33,66,39,23,56])
sub=["maths","science","english","telugu","social"]
plt.plot(sub,student1,label="student1")  #without label we cannot make the legend
plt.plot(sub,student2,label="student2")
plt.title("marks of subjects") #title of plot
plt.legend() #put paranthesis after legend for label appearance
```

```python
student1=np.array([70,45,20,30,60])
student2=np.array([33,66,39,23,56])
sub=["maths","science","english","telugu","social"]
plt.subplot(2,1,1)
plt.plot(sub,student1,label="student1",color="b")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,student2,label="student2",color="k")
plt.legend()
```

```python
A=np.array([230,560,780,127,128])
print("revenue A",A)
B=np.array([200,160,270,127,400])
print("revenue B",B)
profA=np.diff(A)
print("profitsA ",profA)
profB=np.diff(B)
print("profitsB ",profB)
year=(["19-20","20-21","21-22","22-23"])
plt.subplot(1,2,1)
plt.bar(year,profA,label="company1",color="k")
plt.legend(loc="best")
plt.subplot(1,2,2)
plt.bar(year,profB,label="company2",color="b")
plt.legend(loc="best")
```
