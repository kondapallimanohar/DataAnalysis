 the salaries of 5 employees in a company is taken as a list.the tax is 10% if the salary is less than or = to 5000 or it is 15% create a new list with tax amounts [67000,45000,89000,34000,50000] list name [(body of if)   (condition)  else  (body of else)iterate]

```python
salary=[67000,45000,89000,34000,50000]
tax=[]
for i in salary:
  if i <= 50000:
    t=i*0.1
    tax.append(t)
  else:
    t=i*0.15
    tax.append(t)
print(tax)
```

```python
salary=[67000,45000,89000,34000,50000]
tax=[i*0.1 if i<=50000 else i*0.15 for i in salary]
print(tax)
```
